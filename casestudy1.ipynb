# 2. Check for the null values present in the dataset.

# 3.Plot a bar graph of male vs female buyers participated in the sales.


```python
slice_cars_data= cars_data[['Buyer Gender']]
slice_cars_data
BuyGender= slice_cars_data['Buyer Gender'].value_counts()
plt.figure(figsize=(7,5))
slice_cars_data['Buyer Gender'].value_counts().plot(kind='bar',width= 0.4, color =['blue','yellow'])
plt.title('Count of Male Vs Female Car Buyers ', fontsize=20)
plt.xlabel('Gender',fontsize=16)
plt.ylabel('Count of Buyers',fontsize=16)
```
